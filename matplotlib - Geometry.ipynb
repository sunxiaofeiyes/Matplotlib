```python
import matplotlib.pyplot as plt
#绘制几何图形主要是在patches模块中，其各个几何绘制函数其实是一个个类，返回的并不是图像，而是一个类的实例
from matplotlib.patches import Circle,Ellipse,Rectangle,Arc,Wedge,Arrow,FancyArrow
fig,ax = plt.subplots(2,3)

ax1 = ax[0,0]
#下面语句生成一个circle实例.(x,y)指圆心坐标，radius指圆的半径。facecolor指圆的内部填充色，edgecolor指边框颜色。ls指线条形状。lw指线条宽度.
circle = Circle((0.5,1),radius=1,fc='none',ec='b',ls='-.',lw=3)
#将实例导入到子图中以显示
ax1.add_patch(circle)
#设置图片的x轴和y轴范围
ax1.set_xlim(-1,2)
ax1.set_ylim(0,2)
ax1.set_title('Circle')
ax1.autoscale_view()

#生成一个椭圆，利用椭圆也可以绘制圆，只需要将高宽相等即可。
ax2 = ax[0,1]
#生成椭圆使用Ellipse函数，Ellipse(xy,w,h,fc,ec,lw,ls)，xy指椭圆中心的坐标，w指椭圆的宽度，h指椭圆的高度,angle指椭圆逆时针旋转的角度，默认为0。
elle = Ellipse((0.5,0.5),2,1,ec='r',fc='none',lw='2',ls="-",angle=30)
ax2.add_patch(elle)
ax2.set_xlim(-1,2)
ax2.set_ylim(-1,2)
ax2.set_title('Ellipse')
ax2.autoscale_view()

#利用rectangle生成矩形
ax3 = ax[1,0]
#Rectangle是生成矩形的类，Rectangle(xy,w,h,angle,**kwargs)。xy指矩形左下角所在的xy坐标。w指矩形宽度，h指矩形高度，w、h可以接受负数，正数默认向右、向上。
rect1 = Rectangle((0,0),3,3,fc='none',ec='g',lw=2,ls=":",angle=0)
ax3.add_patch(rect1)
ax3.set_xlim(-1,4)
ax3.set_ylim(-1,4)
ax3.set_title('Rectangle')
ax3.autoscale_view()

#利用Arc生成圆弧线
ax4 = ax[1,1]
#Arc(xy,w,h,angle,theta1,theta2),xy指圆心坐标。w、h分别指宽度、高度。angle指圆心逆时针旋转的角度，关键的是theta1和theta2，分别表示圆弧起点的角度和终点的角度，默认
#分别为0和360，即使用Arc不赋值给theta1和theta2时会生成一个圆或一个椭圆。
arc1 = Arc((0,0),1,1,angle=120,theta1=30,theta2=330,ec='black',lw=2,ls='-')
ax4.add_patch(arc1)
ax4.set_xlim(-1,1)
ax4.set_ylim(-1,1)
ax4.set_title('Arc')
ax4.autoscale_view()

#利用wedges生成楔形
ax5 = ax[0,2]
#可以将楔形看成是填充了颜色的圆弧，其函数参数和公式与arc几乎一样。Wedge(xy,r,theta1,theta2)
wedge1 = Wedge((0,0),1,theta1=30,theta2=330,ec='black',fc='g',lw=2)
ax5.add_patch(wedge1)
ax5.set_xlim(-2,2)
ax5.set_ylim(-2,2)
ax5.set_title('Wedge')
ax5.autoscale_view()

#生成箭头
ax6 = ax[1,2]
#第一种办法是使用Arrow类，Arrow(x,y,dx,dy,width,**kwargs)，xy指其实坐标位，dx指箭头横向距离，dy指箭头纵向距离，width指箭头宽度。这个类不能自定义箭头头部
#arrow1 = Arrow(-0.5,-0.5,dx=1,dy=1,width=0.2)
#第二种办法是使用FancyArrow类，相对于Arrow类，其增加了head_width，head_length，shape，可以自己定义箭头头部的长度、宽度和形状。
arrow1 = FancyArrow(-0.5,-0.5,dx=1,dy=1,width=0.1,head_width=0.5,head_length=0.5,shape='full')
ax6.add_patch(arrow1)
ax6.set_xlim(-1,1)
ax6.set_ylim(-1,1)
ax6.set_title('Arrow')
fig.subplots_adjust(wspace=0.5,hspace=0.5)
ax6.autoscale_view()
plt.show()
```
