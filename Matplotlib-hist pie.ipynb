```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.random.randint(0,100,100)
x2 = np.random.randint(0,100,100)
x=[x1,x2]
labels=['A','B']
plt.hist(x,color=['g','r'],histtype='bar',label=labels)
plt.legend()
plt.show()
```

```python
x = [0.2,0.3,0.4,0.05,0.05]
plt.pie(x,explode=(0.05,0.05,0.05,0.05,0.05),labels=['A','B','C','D','E'],autopct='%3.1f%%',pctdistance=0.7,shadow=True,labeldistance=1.1,startangle=60)
plt.show()
```

```python
x = np.random.randn(1000)
plt.boxplot(x,notch=True,vert=False,patch_artist=True,sym='+',widths=0.2)
plt.title('Test')
plt.show()
```
