`````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y = x**3
y1 = x**2
y2 = x*2
plt.plot(x,y,label='y=x**3')
plt.plot(x,y1,label='y=x**2')
plt.plot(x,y2,label='y=x*2')
'''
图例即label，在绘图函数中定义。一个图上往往有多个不同图线，为每个图都添加图例是必须的。label只定义图例的文字，图例的样式要通过legend函数定义。
legend函数能将图中非绘图函数本身的其他内容显示出来。其常用参数有plt.legend(loc,bbox_to_anchor,facecolor,edgecolor,title,shadow,fancybox,fontsize,title_fontsize,**kwargs)
loc,指图例的位置。可选有：'best'，``'upper left', 'upper right', 'lower left', 'lower right'````'upper center', 'lower center', 'center left', 'center right'``
bbox_to_anchor，指线框的位置，接受四元元组(x, y, width, height)。也可接受二元参数(x,y)。其坐标系统是相对于图像的，即百分比形式的坐标系统。
ncol，指图例内部的列数，默认一列
fontsize，图例字体大小
title，图例标题
shadow，是否设置文本阴影，接受True或False
fancybox，是否设置边框为圆角或直角，接受True或False
图像大标题用plt.title()函数定义。参数有plt.title(label,loc,fontdict,pad)。label即标题文字内容，loc指标题位置。fontdict为一个字典，定义
fontsize,fontweight,color,family等。pad指边框。
'''
plt.legend(loc='upper left',title='Test',ncol=3,shadow=True,fancybox=True,title_fontsize=12)

#定义标题，有多种定义方式，见如下代码
plt.title('Test1',loc='left',fontdict={'fontsize':18,'fontweight':2,'color':'r','family':'Times New Roman'})
plt.title('Test2',loc='center')
plt.title('test3',loc='right',color='r',fontsize=18,style='oblique')

plt.show()
`````

`````python
#在以往的代码和学习中，对于文字样式的定义，每一次在函数中都要重新定义，这样子十分麻烦。可以创建一个字典专门用来存储字体样式，在所有函数中都可以直接调用，使得代码十分简洁
font = dict(size=12,family='serif',color='navy',weight='black',style='italic')
plt.plot(x,y)
#有fontdict参数的，可以直接传递font给之。没有的可以使用**font将预先定义好的font样式传递给函数
plt.title('Test',fontdict=font)
plt.xlabel('x numbers',**font)
plt.ylabel('y=x**3',**font)
plt.show()
`````

`````python
#实际上，还可以通过直接设置rcParams属性字典设置字体属性值和文本属性值，这种方式设置的是应用于全局的。
#如果全局都是用一种字体样式可以通过font字典预定义或属性字典修改，如果每一个元素所使用的字体样式不一样则要在函数关键词参数中修改。
#一般来讲每种元素的字体是不一样的，所以推荐在关键词内修改。
`````
