```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y = x*2
plt.plot(x,y)
#annotate(text,xy,xytext,xycoords,textcoords,weight,arrowprops,color,**kwargs)
#text指要注解的文字内容。​xy指被注解内容的坐标位置。xycoords指坐标位置参照系统。xytext指注解文本所处的坐标位置，textcoords指注解文本坐标参照系统。​weight、color指注解文本的字体显示和颜色，arrowprops指箭头属性。​
arrowprops = dict(arrowstyle='-|>',connectionstyle='angle,angleA=0,angleB=90,rad=10',color='g')
bbox = dict(boxstyle='round',fc = 'w',ec='g')
#plt.annotate('medium',xy=(5,10),xycoords='data',xytext=(6,5),textcoords='data',weight='bold',color='r',arrowprops=arrowprops,bbox=bbox)
bbox = dict(boxstyle='circle',fc='black',ec='g',alpha=0.3)
plt.text(1,15,'y=2x',fontsize=14,color='w',bbox=bbox)
plt.text(7,5,'hahaha',fontsize=28,color='g',style='italic',family='Fangsong',alpha=0.4,rotation=30)
plt.text(2,10,'Matplotlib',fontsize=40,color='r',alpha=0.5)
plt.show()
```

```python
x = np.linspace(0,20,1000)
y = np.sin(x)
plt.fill(x,y,c='blue',alpha=0.4)
plt.plot(x,y)
```

```python
x = np.linspace(0,20,1000)
y = np.sin(x)
y1 = np.cos(x)
plt.fill_between(x,y,y1,where=y<=y1,facecolor='r',alpha=0.4)
plt.fill_between(x,y,y1,where=y>y1,facecolor='g',alpha=0.3)
plt.plot(x,y,label='sin')
plt.plot(x,y1,label='cos')
plt.legend(loc='lower left',fontsize=8)
plt.show()
```
