```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y = np.sin(x)
plt.plot(x,y,ls='-',lw=1,label='lines',c='g')  

'''
plot函数用于绘制线图，反应变量间趋势关系时会经常使用。其中，x轴数据和y轴数据必须传递，如果只传递一个数据，则默认将之认为是y轴数据。
ls参数是线条形状，可以传递许多符号参数，如'>'等。
lw即linewidth的缩写，指线条宽度，label参数是指根据x\y数据画出的这条线的名称标签。c是color的缩写，指线条的颜色，可接受颜色的缩写，如red写为r等。
'''
plt.xlabel('X slim',fontsize=12)  #xlabel接受一个字符串，指x轴的标签名。fontsize是指标签的字体大小(下同)
plt.ylabel('Y slim',fontsize=12)  #ylabel接受一个字符串，指y轴的标签名。
plt.title('Sample pic',fontsize=24)  #title指图片的标题，显示在图片正中上方
plt.tick_params(axis='both',labelsize=8)  #tick_params可用来改变轴刻度文字的大小。axis表示要更改哪个轴，labelsize表示更改后的大小。
plt.legend()           #legend()函数能将label标签显示在图片上
plt.xlim(1,8)          #xlim接受两个数值，表示x轴刻度的数值范围。
plt.ylim(0,1)          #ylim接受两个数值，表示y轴刻度的数值范围。
plt.figure(dpi=128,figsize=(10,6))
#plt.show()             #将图表显示在当前程序运行结果下
plt.savefig('sample1.png',bbox_inches='tight')   #使用savefig保存图像到本地。bbox_inches传递tight参数可以使图片周边空白区域消除。
```

```python
x = np.linspace(0.05,10,1000)
y = np.random.randn(1000)
plt.scatter(x,y,label='scatter',c=y,cmap=plt.cm.Reds,edgecolor='none')
plt.legend()
'''
scatter用来绘制散点图，多用来发现变量之间潜在的关系。参数以及图表元素函数多数和plot是一样的，不再详述。其中edgecolor是指点的轮廓颜色，可以传递其他数值，默认为none。
着重要说的是函数映射，可以发现，c既可以作为颜色参数传递颜色，也可以作为颜色映射的参数接受一个轴上的数值数据。例中我传递了y轴数据作为映射，使用
cmap对y轴数据进行映射，使其颜色深浅随数值大小而变化。Reds指颜色，还可以接受Blues，Greens等。具体接受何种颜色可以参照matplotlib的文档。
'''
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,8,9,7,10]
y1 = [3,4,6,1,2]
plt.bar(x,y,width=0.5,align='center',color='g',tick_label=['a','b','c','d','e'],alpha=0.5,label='A')
plt.bar(x,y1,width=0.5,align='center',bottom=y,color='r',alpha=0.5,label='B')
plt.xlabel('X lim')
plt.ylabel('Y lim')
plt.title('Test bar')
plt.legend()
plt.grid(True,axis='y',ls=':',color='b',alpha=0.3)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,8,9,7,10]
y1 = [3,4,6,1,2]
plt.barh(x,y,height=0.5,align='center',color='g',tick_label=['a','b','c','d','e'],alpha=0.5,label='A')
plt.barh(x,y1,height=0.5,align='center',left=y,color='r',alpha=0.5,label='B')
plt.xlabel('Y lim')
plt.ylabel('X lim')
plt.legend()
plt.grid(True,axis='x',ls=':',color='b',alpha=0.3)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = [6,8,9,7,10]
y1 = [3,4,6,1,2]
bar_width=0.3
tick_label=['a','b','c','d','e']
plt.bar(x,y,width=0.3,align='center',color='g',alpha=0.5,label='A')
plt.bar(x+bar_width,y1,width=0.3,align='center',color='r',alpha=0.5,label='B')
plt.xticks(x+bar_width/2,tick_label)
plt.legend()
plt.show()
```
