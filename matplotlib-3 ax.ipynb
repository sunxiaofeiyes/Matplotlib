```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(100)
y = np.sin(x)
y1 = np.cos(x)
#使用subplot创建一个画布，根据subplot(C,R,P),该画布为1行2列的画布，在1号子区上作画
plt.subplot(1,2,1,projection='polar')
plt.scatter(x,y)
#显示2号区，在2号子区作画
plt.subplot(1,2,2)
plt.scatter(x,y1)
plt.show()
```

```python
#使用subplot分不同区域作画
#先用subplot创建一个2行1列的1区画布并作画
plt.subplot(2,1,1)
plt.scatter(x,y)
#使用subplot再创建一个2行2列的画布，在3区作画。因为其第一行已被占据为重叠空间，所以第一行不显示分列了
plt.subplot(2,2,3)
plt.scatter(x,y1)
#在4区作画
plt.subplot(2,2,4)
plt.scatter(x,y)
plt.show()
```

```python
#使用subplot2grid()绘制子图
#在第一行第一列绘制
plt.subplot2grid((2,3),(0,0))
plt.scatter(x,y)
#在第一行第二列绘制一个宽度为2的图
plt.subplot2grid((2,3),(0,1),colspan=2)
plt.scatter(x,y1)
#在第二行第1列绘制一个宽度为3的图
plt.subplot2grid((2,3),(1,0),colspan=3)
plt.scatter(x,y)
plt.show()
```

```python
#使用subplots函数绘图
#先创建一个绘图画布，网格形状为2行3列。返回的为元组形式，fig指整个画布。ax指分区集合。
fig, ax = plt.subplots(2,3)
#调用第一个分区，ax可看作是个ndarray，第一个数字表示行，第二个数字表示列，索引从0开始，依此类推。
ax1 = ax[0,0]
ax1.scatter(x,y)
#使用set_title设置子图1的标题
ax1.set_title('pic 1')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')

#调用第二个分区，第一行第二列。
ax2 = ax[0,1]
ax2.scatter(x,y1)
ax2.set_title('Pic 2')
ax2.set_xlim(0,2)

#调用最后一个分区，第2行第3列
ax3 = ax[1,2]
ax3.scatter(x,y)
#确定整个画布的标题
plt.suptitle('Test')
plt.show()
```

```python
#共享坐标轴
x = np.random.randn(1000)
x1 = np.arange(100)
x2 = np.arange(1000)
x3 = np.arange(100)
y = np.sin(x)
y1 = x1*3
y2 = x2/100
y3 = x3-1
fig,ax = plt.subplots(2,2)
ax1 = ax[0,0]
ax1.scatter(x,y)
ax2 = ax[0,1]
ax2.plot(x1,y1)
ax3 = ax[1,0]
ax3.plot(x2,y2)
ax4 = ax[1,1]
ax4.plot(x3,y3)
fig.subplots_adjust(wspace=0,hspace=0)
plt.show()
```
